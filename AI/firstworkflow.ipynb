```python
from dotenv import load_dotenv
load_dotenv(override=True) 
```

```python
import os
```

```python
openai_api_key = os.getenv("API_TOKEN") 
```

```python
if openai_api_key:
    print(f"OpenAI Key exists and begins with: {openai_api_key[:10]}....")
else:
     print(f"OpenAI Key not found, please set the API_TOKEN environment variable") 
```

```python
from openai import OpenAI
openai_python_client = OpenAI(api_key=openai_api_key, base_url="https://openrouter.ai/api/v1")
print("OpenAI Client created", openai_python_client)
```

```python
from IPython.display import Markdown, display
question = "Who hosted 2022 world cup?"
messages=[
    {"role": "user", "content": question}
]
response = openai_python_client.chat.completions.create(
  model="openai/gpt-5-mini",
  messages=messages,
  stream=False,
)
display(Markdown(response.choices[0].message.content)) 
```
